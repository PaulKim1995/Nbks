# Is a solar signal embedded in a century of synoptic wind data?

This iPython notebook reads in daily HB synoptic data, converts it into monthly counts by wind direction, and then composite by solar cycle occurence with MC-significance testing.

*Benjamin A. Laken, June 2014*

```python
# Import required modules
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# ... Compile the custom functions

#......
 # Function to generate strings of 'YYYY-MM' to call pd object
def unq_mnths(data): 
    mn_dt = min(data.index).date()
    mx_dt = max(data.index).date()
    smn = mn_dt.month       # Start month
    syr = mn_dt.year        # Start year
    emn = mx_dt.month       # End month
    eyr = mx_dt.year        # End year
    outs = []               # list to append output
    out_dt =[]              # Monthly datetime list
    while((syr != eyr) | (smn != emn)):
        outs.append(str(syr)+'-'+str(smn))
        out_dt.append(pd.datetime(syr,smn,1).date())
        smn = smn +1
        if(smn == 13):
            smn = 1
            syr = syr + 1
    print 'Index subscript, unique YYYY-MM, of n =', len(outs)
    return outs,out_dt


#......
# Calculate days per month for a wind direction
# dtindx is the YYYY-MM formated string output (made by unq_mnths function above)
# data is the PD Series object (with the date as an index)
# Compass is a string of: N, NE, E, SE, S, SW, W, NW
def blowy(data,dtindx,compass):
    Ncnt=[] # Number of days of a given wind direction per month (int)
    Dpm = []   # Number of days in each month (int)
    for i,n in enumerate(dtindx):
        tmp = []
        tmp = data[n]
        tmp = np.array(tmp)
        Dpm.append(len(tmp))                   # Days in the month
        Neles = np.where(tmp == compass)
        Ncnt.append(len(Neles[0]))             # Wind Days of a direction per month
    Dpm = np.array(Dpm)
    Ncnt = np.array(Ncnt)
    return Ncnt,Dpm


# ====================== Seasonal (climatology) ================================
# Identify the climatology by calander months.
#
# Input: mnth(list of months as integers from 1 to 12)
#        dat(data corresponding to months)
#        note - mnth and dat should be the same lenght
#
#Output: result - a 12 x 3 array (for every month a mean,
#        a standard deviation, and a count is outputted).
#
#
def Seasonal(mnth,dat):
    tot = len(dat)    # how many elements in the data arrays
    tot2 = len(mnth)
    if tot != tot2:
        print 'Error: months and data input are not of equal lenght.'
    result = np.zeros(shape=(12,3))
    # entry format will be for each of the 12 (months), average, stdev, count
    for i in xrange(12):
        lookup = np.where(mnth == i+1)
        aaa = 0
        aaa = dat[lookup[0]]
        result[i,2] = len(aaa)              # Number of months
        result[i,1] = np.std(aaa)           # Standard deviation of data
        result[i,0] = np.average(aaa)
    return result
#......

# ======================SEM RANGE===================================
# Input the mean, standard deviation, and count
# Return the standard error of the mean, 
#and (SEM + mean), (SEM - mean)
def SEM_range(mn,stdev,cnt):
    nlen = len(mn)
    result = np.zeros(shape=(nlen,3),dtype='f')
    for n in xrange(nlen):
        result[n,0]  = stdev[n]/(cnt[n] **(0.5))
        result[n,1]  = mn[n] + result[n,0]
        result[n,2]  = mn[n] - result[n,0]
    return result
#----------------------------------------------------------------
```

###HB-synoptic types classified by their wind direction 

Asccii code assigned to synoptic types, and its relation to the direction of wind. 

|Direction | H-B types
| ------------- |:-------------:| -----:|
|N  | 'NA','NZ','HNA','HNZ','HB','TRM'|
|NE | 'NEA','NEZ'|
|E | 'HFA','HFZ','HNFA','HNFZ'|
|SE |'SEA','SEZ'|
|S  |'SA','SZ','TB','TRW'|
|SW |'SWA','SWZ'|
|W  |'WZ','WS','WA','WW'|
|NW |'NWA','NWZ'|

```python
# Create a dictionary object to define which wind direction the HB-types relate to.
wind_dict={'NA':'N','NZ':'N','HNA':'N','HNZ':'N','HB':'N',
           'TRM':'N','NEA':'NE','NEZ':'NE','HFA':'E',
           'HFZ':'E','HNFA':'E','HNFZ':'E','SEA':'SE','SEZ':'SE',
           'SA':'S','SZ':'S','TB':'S','TRW':'S','SWA':'SW','SWZ':'SW',
           'WZ':'W','WS':'W','WA':'W','WW':'W','NWA':'NW','NWZ':'NW',
           'HM':0,'TM':0,'U':0,'BM':0}

direction = np.zeros(len(hb),dtype='string')
for i,n in enumerate(hb):
    direction[i] = wind_dict[n]
    #print i,'-',n,':',wind_dict[n]
```

```python
dates = pd.date_range('18810101', '20001231', freq='D')    # Create a date object
df = pd.Series(direction,index=dates)   # PD Series (of the wind direction by day)
```

```python
a =  min(df.index).date()
b = max(df.index).date()
print a.month,a.year
print b.month,b.year
```

```python
# Generate a list of unique 'YYYY-MM' strings from the input of daily data
outlist,out_dt = unq_mnths(df)

winds = ['N','NE','E','SE','S','SW','W','N']   # String list of wind directions
monthly_counts={}

for n,i in enumerate(winds):
    print 'Creating counts per month for ',i
    tmpmonth = []
    tmpDcnt = []
    tmpmonth,tmpDcnt = blowy(df,outlist,i)
    monthly_counts[str(i)]=tmpmonth
```
