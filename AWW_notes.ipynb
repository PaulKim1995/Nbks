```python
import pandas as pd                   # Import PANDAS
import datetime as dt                 # We will also use a function from datetime
import matplotlib.pyplot as plt
from numpy import linalg              # Needed for linear fitting           
from scipy import stats               # Needed for Pearson's r 

co2 = pd.read_csv('/Users/Ben/Work/Python/Data/co2.txt',header=0)
Nhemi = pd.read_csv('/Users/Ben/Work/Python/Data/NHemi_NASA.csv',header=0)
Shemi = pd.read_csv('/Users/Ben/Work/Python/Data/SHemi_NASA.csv',header=0)

co2.index = co2.Year      # Make the year integers the index
Nhemi.index = Nhemi.Year
Shemi.index = Shemi.Year
```

```python
# Merge the dataframes, stripping the data to allign with left (CO2) indexes.
all = pd.merge(co2,Nhemi,how='left')
all = pd.merge(all,Shemi,how='left')
all.index = all.Year
```

### Pearsons correlation from the variables using PANDAS function

```python
cout = all.corr(method='pearson',min_periods=1)  # Save the data structure to 'cout'
#all.corr(method='pearson',min_periods=1)         # Also, just display it
print 'Annual CO2 vs Nhemi r =',cout.PPM.Nhemi
print 'Annual CO2 vs Shemi r =',cout.PPM.Shemi
```

### Create 10-year running means, and check the Pearson's r value

```python
ppm10 = pd.stats.moments.rolling_mean(all.PPM,window=10,center=True)
shemi10 = pd.stats.moments.rolling_mean(all.Shemi,window=10,center=True)
nhemi10 = pd.stats.moments.rolling_mean(all.Nhemi,window=10,center=True)

# Also, create a dataframe out of the smooths..
smth10 = {'PPM10':ppm10,'Shemi10':shemi10,'Nhemi10':nhemi10}
smooth = pd.DataFrame(smth10,index=all.index)
```

```python
cout10 = smooth.corr(method='pearson',min_periods=1)

print 'Decadal CO2 vs Nhemi r =',cout10.PPM10.Nhemi10
print 'Decadal CO2 vs Shemi r =',cout10.PPM10.Shemi10
```

## Calculate $\Delta$ values: the year-to-year change

```python
def delta(datin):
    tmp = []
    tmp = np.array(datin)
    tmp_out =[]
    cnt = 0
    for i  in tmp:
        if(cnt >= 1):
            tmp_out.append((tmp[cnt] - tmp[cnt-1]))
        cnt = cnt+1
    return np.array(tmp_out)
```

```python
Nhemi_delta = delta(all.Nhemi)
Shemi_delta = delta(all.Shemi)
PPM_delta = delta(all.PPM)
```

Plot the $\Delta$ Temp against $\Delta$ CO$_{2}$

```python
x = PPM_delta
y1 = Nhemi_delta
y2 = Shemi_delta

A = np.vstack([x, np.ones(len(x))]).T

w_1 = linalg.lstsq(A,y1)[0] 
w_2 = linalg.lstsq(A,y2)[0]

Nhemi_dfit = w_1[0]*x+w_1[1] # regression line
Shemi_dfit = w_2[0]*x+w_2[1] # regression line
```

### Conclusion

It is true that the relationship (slope) between annual CO$_{2}$ and hemispheric temperature is smaller than the slope between $\Delta$ CO$_{2}$ and temperature, by roughly one tenth. 

This would, at first glance suggest that the year-to-year changes in CO$_{2}$ have more impact on hemispheric temperature. However, the correlation between the data is much lower for the $\Delta$ values – indicating that the slope obtained from the latter is far less significant.
