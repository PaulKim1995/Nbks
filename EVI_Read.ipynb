## Plot a time-series

```python
# Plot time series 

my_ts = plt.figure()
my_ts.set_size_inches(10,5)        # Specify the output size
ax1 = my_ts.add_subplot(111)        # Add an axis frame object to the plot (i.e. a pannel)


ax1.plot(MODIS.index.date,MODIS.GT_mean,'bx')
ax1.plot(LS5.index.date,LS5.GT_mean,'rx')
ax1.plot(LS7.index,LS7.GT_mean,'r.')


ax1.set_title(r'Grass-type EVI values from Landsat (5 = red & 7 = green) and MODIS(blue)')
ax1.set_ylabel(r'EVI ($\mu$)')
ax1.set_xlabel('Date')
ax1.grid(True)

plt.show(my_ts)
#my_ts.savefig('Monthly_ts.pdf',dpi=300)
```

## Example of how to scatter plot with error bars, and fit considering error

```python
from scipy import odr    # http://docs.scipy.org/doc/scipy-0.14.0/reference/odr.html
from scipy.stats import linregress  # Also, just a simple linear regression function
```

```python
#First calculate a simple linear regression for the data...
xvals = np.arange(0.1,0.9,0.1)   
vals = linregress(MODIS.GT_mean[0:20],MODIS.FT_mean[0:20])
yvals_noerr =[i* vals[0] + vals[1]for i in xvals]
```

```python
# Now use the Orthogonal distance regression module 
# Set example of real data with real x, and y errors 
mydata = odr.RealData(MODIS.GT_mean[0:20],MODIS.FT_mean[0:20],sx=MODIS.FT_std[0:20],
             sy=MODIS.GT_std[0:20])

def f(B, x):
    '''Linear function y = m*x + b'''
    # B is a vector of the parameters.
    # x is an array of the current x values.
    # x is in the same format as the x passed to Data or RealData.
    #
    # Return an array in the same format as y passed to Data or RealData.
    return B[0]*x + B[1]

linear = odr.Model(f)
myodr = odr.ODR(mydata, linear, beta0=[1., 2.]) 

myoutput = myodr.run()
myoutput.pprint()
```

```python
print myoutput.beta   # Orthogonal distance regression output
yvals_werr = [i * myoutput.beta[0]+ myoutput.beta[1] for i in xvals] # Create y values for liner model
```

```python
# Plot it all
my_scatter = plt.figure()
my_scatter.set_size_inches(6,6) 
ax1 = my_scatter.add_subplot(111)        

ax1.errorbar(MODIS.GT_mean[0:20],MODIS.FT_mean[0:20],yerr=MODIS.FT_std[0:20],
             xerr=MODIS.GT_std[0:20],c='r',fmt='o',alpha=0.7)

ax1.plot(xvals,yvals_werr,'b-')  # Solid blue = Orthogonal distance regression
ax1.plot(xvals,yvals_noerr,'g-') # Green = Simple regression, no error

ax1.set_xlim(0.2,0.7)
ax1.set_ylim(0.2,0.7)

ax1.set_ylabel('Grass Type EVI')
ax1.set_xlabel('Forest Type EVI')
ax1.grid(True)
ax1.set_title('FT/GT cross-correlation and two-types of lin. regression \n'+
              'Blue = ODR Lin. Reg. considering error \n'+ 
              'Green = Simple Lin. Reg.')
plt.show(my_scatter)
```
