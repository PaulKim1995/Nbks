```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Read the data from a text file
with open('/Users/Ben/Work/Python/Data/HB_data.txt') as f:
    lines = (line for line in f if not line.startswith('#'))
    data = np.loadtxt(lines, skiprows=1, dtype='str')
hb = np.array(data[:,3])
```

```python
# Create a dictionary object to define which wind direction the HB-types relate to.
'''
N  = np.array(['NA','NZ','HNA','HNZ','HB','TRM'])
NE = np.array(['NEA','NEZ'])
E  = np.array(['HFA','HFZ','HNFA','HNFZ'])
SE = np.array(['SEA','SEZ'])
S  = np.array(['SA','SZ','TB','TRW'])
SW = np.array(['SWA','SWZ'])
W  = np.array(['WZ','WS','WA','WW'])
NW = np.array(['NWA','NWZ'])
'''
# Use the dictionary to define the synoptic type wind direction
wind_dict={'NA':'N','NZ':'N','HNA':'N','HNZ':'N','HB':'N',
           'TRM':'N','NEA':'NE','NEZ':'NE','HFA':'E',
           'HFZ':'E','HNFA':'E','HNFZ':'E','SEA':'SE','SEZ':'SE',
           'SA':'S','SZ':'S','TB':'S','TRW':'S','SWA':'SW','SWZ':'SW',
           'WZ':'W','WS':'W','WA':'W','WW':'W','NWA':'NW','NWZ':'NW',
           'HM':0,'TM':0,'U':0,'BM':0}

direction = np.zeros(len(hb),dtype='string')
for i,n in enumerate(hb):
    direction[i] = wind_dict[n]
    #print i,'-',n,':',wind_dict[n]
```

```python
dates = pd.date_range('18810101', '20001231', freq='D')    # Create a date object
df = pd.Series(direction,index=dates)   # PD Series (of the wind direction by day)
```

```python
#......
def unq_mnths(data):  # Function to generate strings of 'YYYY-MM' to call pd object
    mn_dt = min(data.index).date()
    mx_dt = max(data.index).date()
    smn = mn_dt.month       # Start month
    syr = mn_dt.year        # Start year
    emn = mx_dt.month       # End month
    eyr = mx_dt.year        # End year
    outs = []               # list to append output
    out_dt =[]              # Monthly datetime list
    while((syr != eyr) | (smn != emn)):
        outs.append(str(syr)+'-'+str(smn))
        out_dt.append(pd.datetime(syr,smn,1).date())
        smn = smn +1
        if(smn == 13):
            smn = 1
            syr = syr + 1
    print 'Index subscript, unique YYYY-MM, of n =', len(outs)
    return outs,out_dt
#......

outlist,out_dt = unq_mnths(df)

#df[outlist[1]]  # to pull out one month for example
```

```python
#----
# Calculate days per month for a wind direction
# dtindx is the YYYY-MM formated string output (made by unq_mnths function above)
# data is the PD Series object (with the date as an index)
# Compass is a string of: N, NE, E, SE, S, SW, W, NW
def blowy(data,dtindx,compass):
    Ncnt=[] # Number of days of a given wind direction per month (int)
    Dpm = []   # Number of days in each month (int)
    for i,n in enumerate(dtindx):
        tmp = []
        tmp = data[n]
        tmp = np.array(tmp)
        Dpm.append(len(tmp))                   # Days in the month
        Neles = np.where(tmp == compass)
        Ncnt.append(len(Neles[0]))             # Wind Days of a direction per month
    Dpm = np.array(Dpm)
    Ncnt = np.array(Ncnt)
    return Ncnt,Dpm
# ----

Direction = 'W'
N_month,Dcnt = blowy(df,outlist,Direction)

#N_month = N_month/Dcnt
```

```python
plt.figure()
plt.plot(out_dt,N_month)
plt.title(Direction+' wind days per month')

#print b
#xxx = np.where(b == 'S')
#print len(xxx[0])

#print df['1881-01-01']
#print df['2000-12-26':'2000-12-31']
#df.count
```
