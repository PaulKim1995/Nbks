### Now, we will use the netCDF4 python package to read one of the files.

In netCDF format, data (variables) are stored with meta-data (descriptions of the data), and dimensions. For gridded Earth science data, the variables are whatever field you are measuring, the metadata contains information about the nature of the measurements, and, often the values of missing data, and scaling values which need to be applied to convert the data in the variable from a short (easy to store) integer, into a real (floating-point) value. The variables are a system of spatio-temporal co-ordinates (e.g. lat, lon, altitude, time).


```python
test1 = Dataset(find_eg[0], 'r')    # This line actually does the reading (using netCDF4 package)
print 'Reading ',find_eg[0]
test1.close
```

```python
for dimobj in test1.dimensions.values():    # Examine the dimensions
    print dimobj
```

```python
for varobj in test1.variables.values():    # Examine the variable
    print varobj
```

Note that the variables contain both the dataset (cloud fraction), and also the values of lat, and lon. The lat and lon are specified as DIMENSIONS of the cloud fraction dataset.

```python
cfrac = test1.variables.get('Cloud_Fraction_Day_Mean')
lon = test1.variables.get('lon')
lat = test1.variables.get('lat')
```

 The netcdf object is a MASKED_ARRAY. This is a numpy class that shows '--' 
 where the -Missing_value is present in the variable. 
 The behaviour of the Masked array is that, it considers elements which as masked to be
 a boolean equal to True. And unmasked elements to be a bool = False. When querying the
 datatype of the array, masked elements will show as a type <np.ma.Masked_array>, while
 unmasked elements will show as whatever datatype the values are.

 Note, It can be tricky to plot an image with masked values. So it is easy to replace them
 with np.nan, or pythons 'None' data type. I note that you cannot directly write them, as
 data read in with netCDF4, (the netcdf objects) are read only.

```python
%%time
dat = np.zeros(cfrac.shape,dtype='f') 
for i in xrange(len(lat)):      # A slow and old-fashioned way to loop through every element
    for j in xrange(len(lon)):
        dat[i,j] = float(cfrac[i,j])
print 'Prepared data'
```

The data can now be plotted simply as a plain raster image as a quick-view

```python
plt.imshow(dat,origin='lower')
colorbar()
```

### Use logic on pressure and $\tau$ to identify pixels of low ($\gt800$ mb) stratocumulus

```python
%%time
# the time magic function above just times the execution of this whole cell. For fun.
msc_msk = np.zeros(cfrac.shape,dtype='f')   # Initilize an array
# A slow and old-fashioned way to loop through every element
for i in xrange(len(lat)):
    for j in xrange(len(lon)):
        if((float(cpress[i,j]) >=800.)&(float(ctau[i,j]) >=3.6)&(float(ctau[i,j]) <= 23.)):
            msc_msk[i,j] = 1.
print 'Identified MSc data'
```

```python
# Show the pixels identifid as low Stratocumulus (Sc)
m = Basemap(projection='cyl', resolution='c',llcrnrlat=-90,urcrnrlat=90,llcrnrlon=-180,urcrnrlon=180)
m.drawcoastlines(linewidth=0.5)
m.drawparallels(np.arange(-90.,91.,60.),labels=[1,1,0,0],fontsize=10)  # the labels indicates which
m.drawmeridians(np.arange(0.,361.,90.),labels=[0,0,1,1],fontsize=10)   # side of the graph to label.
m.pcolormesh(lon[:],lat[:],msc_msk,cmap=cm.bwr)

print 'Red = low Sc, Blue = not low Sc'
```

### Area-averaging

Before averaging the data in space, we will need to weight the values by their true area.  
    
The formulae to calculate the area is given at the following link: http://badc.nerc.ac.uk/help/coordinates/cell-surf-area.html

Fear not. In the below cell, I have already written this formulae into a nice friendly python function. All you have to do is call the function, specifying the number of latitude and longitude grid cells (as a single integer value each). The function will then return two grids. One is the actual area (in km$^{2}$) of each grid cell, the other is the fractional weight each cell has (a scale factor). 

To create an equal area average, each cell must contribute a change relative to its area. All you need to do to account for this is to multiply whatever value each grid cell of data has by the scale factor when calculating an area average.

```python
def eq2act_area(lat,lon):
    radlon1 = np.zeros(lon,dtype='f')
    radlon2 = np.zeros(lon,dtype='f')
    sinlat1 = np.zeros(lat,dtype='f')
    sinlat2 = np.zeros(lat,dtype='f')
    surface_area = np.zeros([lat,lon],dtype='f')
    factor_surface = np.zeros([lat,lon],dtype='f')
    earth_radius=6367.4447              # This value is from Wolfram Alpha
    radian_conversion = np.pi/180.0

    #At the specified resolution how many degrees per a pixel?
    pixlon=360./float(lon)
    pixlat=180./float(lat)
    print 'Grid resolution is',pixlon,'x',pixlat,' (Units = Degree)'
    
    for j in xrange(lon):
        radlon1[j] = (((j+1) * pixlon) - pixlon)  * radian_conversion
        radlon2[j] = ((j+1) * pixlon) * radian_conversion

    for i in xrange(lat):
        sinlat1[i]= sin(((((i+1)*pixlat)-pixlat)-90.0) * radian_conversion)
        sinlat2[i]= sin((((i+1)*pixlat)-90.) * radian_conversion)
    
    for i in xrange(lat):
        for j in xrange(lon):
            surface_area[i,j]=(earth_radius*earth_radius)*(radlon2[j]-radlon1[j])*(sinlat2[i]-sinlat1[i]) 

    # Check everything is okay by calculating the Earth's surface area.
    total_area=0.0   
    for i in xrange(lat):
        for j in xrange(lon):
            total_area=total_area+surface_area[i,j]

    print 'Calculated area of Earth =',total_area,'km'
    print 'Actual value should be 5.1x10^8 km^2 '
    
    # Crucially, create an array to use to weight gridded data      
    for i in xrange(lat):
        for j in xrange(lon):
            factor_surface[i,j]=surface_area[i,j]/surface_area[lat/2,lon/2]
    print 'Center of grid = ',surface_area[lat/2,lon/2],'km^2 : weight = ',factor_surface[lat/2,lon/2]
    print 'Edge of grid = ',surface_area[0,lon/2],'km^2 : weight = ',factor_surface[0,lon/2]
    return surface_area,factor_surface
```

```python
area,weight = eq2act_area(180,360)
```

```python
# The area of each grid cell in km^2
plt.imshow(area)
plt.title(r'Area of a 1$\times$1$^{\degree{}}$ grid cell (km$^{2}$)')
colorbar()
```

```python
plt.imshow(weight)
plt.title(r'The surface factor for each grid cell ')
colorbar()
```

### Final step: applying the masks and surface weights to calculate area-averages

```python
%%time

# top.mask = False over land, True over Ocean
# msc_msk[i,j] = 1 where Sc cloud, 0 where not.

msc_cfrac = []  # These will be lists, to which we can append
msc_ccn = []    # all the weighted pixel values which meet the logic test.
msc_radi =[]

for i in xrange(len(lat)):
    for j in xrange(len(lon)):
        # Identify the condition of being Sc & over ocean
        if((top.mask[i,j] == True)&(msc_msk[i,j] == 1.)): 
            msc_cfrac.append(cfrac[i,j] * weight[i,j])
            
print 'Of ',360*180,' pixels, ',len(msc_cfrac),' were MSc type: i.e. ',float(len(msc_cfrac))/(360*180) *100,'%'
```

Now you have lists of weighted values from wherever MSc cloud types occurred on a given day. From this, you can easily calculate means, and SEM values.

```python
print ' MSc-type cloud fraction on',greg_day
print 'Mean & Error:', np.average(msc_cfrac),'±',np.std(msc_cfrac)/np.sqrt(len(msc_cfrac))
```
