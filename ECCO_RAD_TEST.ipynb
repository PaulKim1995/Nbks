## EUR-11 Model Check

From outputs of upwelling radiation and surface temp Check emissivity is balanced in the EUR-11 models: 

$\epsilon = \frac{rlus}{\sigma T_{s}^{4}}$

```python
from scipy import constants
import matplotlib
import numpy as np
import matplotlib.pyplot as plt
pylab.rcParams['figure.figsize']=(10.0,8.0) 
```

```python
print constants.sigma   # Use the Stefan-Boltzman constant from Scipy
```

### Read the data from zipped files

```python
pth = '/mn/vann/metos-d1/blaken/ECCO/Outputs/EUR-11_ICHEC-EC-EARTH_r1i1p1_KNMI-RACMO22E'\
                  +'_v1_day/15/1559/155980/'

rlus = np.loadtxt(pth+'rlus_rcp45_20310101-20351231txt.gz',skiprows=1,dtype=float)
tas = np.loadtxt(pth+'tas_rcp45_20310101-20351231txt.gz',skiprows=1,dtype=float)
```

### Test disply the data

```python
plt.plot(rlus,tas,'r.')
plt.xlabel(r'Upwelling LW radiation (W/m$^{2}$)')
plt.ylabel(r'Near-surface air temp. ($^{o}C$)')
plt.grid(True)
```

### Calculate emissivity ratio over 5-years of daily data from a random EUR-11 file

```python
emiss = rlus/(constants.sigma * tas **4)
```

```python
plt.plot(emiss)
plt.ylabel(r"emissivity ratio")
plt.grid(True)
```
